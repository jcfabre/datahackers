#  Histogram with matplotlib and seaborn

### Imports

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Matplotlib

#### 1. With axes

```python
data = np.random.binomial(100, 0.55, 1000)

fig, ax = plt.subplots()
ax.hist(data, bins = 30)  # bins represents the number of bars
```

#### 2. With plt.hist()

```python
data = np.random.binomial(100, 0.55, 1000)
plt.hist(data, bins = 30)
```

### Seaborn

```python
data = np.random.binomial(100, 0.55, 1000)
sns.distplot(data)
```

```python
data = np.random.binomial(100, 0.55, 1000)
sns.distplot(data, kde = False, rug = True)
```

```python
sns.distplot(data, hist = False, rug = True)
```
