# Plots in a grid with matplotlib

### Imports

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Organize plots within a grid

#### __Create a 2 x 2 grid figure with 4 subplots__

   __ 1. First way __

We create a figure object and add axis one at a time

```python
fig = plt.figure(figsize = (15,7))

for i in range(1,5):
    x = np.linspace(0,100,120)
    y = np.random.randn(120)
    ax = fig.add_subplot(2,2,i)
    ax.plot(x,y)
    ax.set_title("Graph {0}".format(i), fontweight='bold')

fig.suptitle("Master graph",  fontsize=14, fontweight='bold')
fig.subplots_adjust(top=0.85)  # add space between figure title and subplots

plt.show()
```

   __ 2. Second way __

We create the figure object and the axis at once

```python
x = np.linspace(0,100,120)
y = np.random.randn(120)

fig = plt.figure(figsize=(15,7))
fig, axarr = plt.subplots(nrows = 2, ncols = 2, figsize = (15,7))

axarr[0, 0].plot(x, y)
axarr[0, 0].set_title('Axis [0,0]')
axarr[0, 1].scatter(x, y)
axarr[0, 1].set_title('Axis [0,1]')
axarr[1, 0].plot(x, y ** 2)
axarr[1, 0].set_title('Axis [1,0]')
axarr[1, 1].scatter(x, y ** 2)
axarr[1, 1].set_title('Axis [1,1]')


fig.suptitle("Master graph",  fontsize=14, fontweight='bold')
fig.subplots_adjust(top=0.85)  # add space between figure title and subplots

plt.show()
```

#### __Create 2 graphs that share the X axis__

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2/3)

fig, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(x, y)
axarr[0].set_title('The 2 graphs share the X axis')
axarr[1].scatter(x, y)
```

#### __Create 2 graphs that share the Y axis __

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig, (ax1, ax2) = plt.subplots(1,2,sharey = True,figsize =(10,5))
ax1.plot(x,y)
ax2.scatter(x,y)
fig.suptitle("Sharing the Y axis", fontsize = 14)
```
