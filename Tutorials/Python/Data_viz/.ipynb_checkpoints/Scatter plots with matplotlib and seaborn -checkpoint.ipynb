#  Scatter plots with matplotlib and seaborn 

### Imports

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
iris = "https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/datasets/iris.csv"

df = pd.read_csv(iris)
df.drop(df.columns[0],axis = 1, inplace = True)
df.head()
```

### With Matplotlib

#### Changing the color

```python
fig, ax = plt.subplots()

#Mapping species with colors
species = df["Species"].unique()
colors = ["red", "blue","green"]
colors_dict = dict(zip(species, colors))

#One at a time
for spec in species:
    ax.scatter(df["Sepal.Length"].where(df.Species == spec), 
               df["Sepal.Width"].where(df.Species == spec), 
               label = spec,
               c = colors_dict[spec])

ax.legend()
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
```

#### Changing the size

```python
fig, ax = plt.subplots()

#Mapping species with colors
species = df["Species"].unique()
colors = ["red", "blue","green"]
colors_dict = dict(zip(species, colors))

#Mapping Petal Length with sizes
petal_sizes = df["Petal.Length"].unique()
plot_sizes = np.linspace(20,50, len(petal_sizes))
sizes_dict = dict(zip(petal_sizes, plot_sizes))
df["Sizes"] = df["Petal.Length"].apply(lambda x: sizes_dict[x])

#One at a time
for spec in species:
    ax.scatter(df["Sepal.Length"].where(df.Species == spec), 
               df["Sepal.Width"].where(df.Species == spec), 
               label = spec,
               s = df["Sizes"], 
               c = colors_dict[spec])

ax.legend()
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
```

### With Seaborn

```python
sns.jointplot(df["Sepal.Length"],df["Sepal.Width"], size = 5)
```

```python
g = sns.FacetGrid(df, hue="Species", size=5)
g = g.map(plt.scatter, "Sepal.Length", "Sepal.Width")
g = g.add_legend()
g
```
