# Sorting rows of a Pandas Dataframe

### Imports

```python
% matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(20,5), columns = list("ABCDE"))
df
```

### Sort one column

```python
df.sort_values(by="A").head(10)
```

### Ascending or Descending order

```python
df.sort_values(by="A", ascending = False).head(10)
```

### Sort multiple columns

```python
df["names"] = ["tom","alex","chris","jack"]*5 
df.head()
```

```python
df.sort_values(["names","A"], ascending = [True, False]).head()
```
