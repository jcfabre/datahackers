# Select rows based on multiple conditions

### Imports

```python
% matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
from StringIO import StringIO
```

### Reading the data into a DataFrame

```python
url = "https://docs.google.com/spreadsheets/d/1sQOtIwkEiTe4kwwRHw5bwi72ZJHXlAVcCNMImr-Ml6U/export?format=csv&gid=0"

data = requests.get(url).content.decode("utf-8")
df = pd.read_csv(StringIO(data))

df.head(3)
```

### Numerical condition

```python
df[df["round"] > 2].head(3)
```

```python
df[df.f1name == "Ronda Rousey"].head(3)
```

### AND condition

```python
df[(df.method == "Submission") & (df.method_d == "Armbar")]
```

### OR condition

```python
df[((df.f1name == "Royce Gracie") & (df.f1result == "win")) | 
   ((df.f2name == "Royce Gracie") & (df.f2result == "win"))].head(3)
```

### Select rows if values of a column ARE in a list

```python
sub_techniques = ["Armbar","Triangle Armbar","Front Choke", "Bulldog Choke"]

df[df.method_d.isin(sub_techniques)]
```

### Select rows if values of a column ARE NOT in a list

```python
sub_techniques = ["Armbar","Triangle Armbar","Front Choke", "Bulldog Choke"]

df[~df.method_d.isin(sub_techniques)].head()
```

### Using boolean to sort rows

#### SImple example: getting the first row of the Dataframe

```python
len(df)
```

```python
bool_list = [True] + ([False] * (len(df)-1))
df[bool_list]
```

#### Using map

```python
df[df["event_place"].map(lambda place: "Las Vegas" in place)].head()
```

#### Using conditions

```python
cond1 = df["method"] == "KO"
cond2 = df["round"] != 1

df[cond1 & cond2]
```
