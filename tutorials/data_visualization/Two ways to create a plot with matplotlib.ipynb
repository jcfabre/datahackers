# Imports

```python
%matplotlib inline

import pandas as pd
import matplotlib
matplotlib.rcParams['backend'] = 'GTKAgg'
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

# Matplotlib

   __1. Non-pythonic (quick) way__

__Uses a global function__

```python
x = np.linspace(0,16,100)
y = x + np.random.randn(100)
temp = pd.DataFrame({"Days": x, "Temps": y})
plt.plot(x, y)
plt.show()
```

__ Adding title and labels __

```python
x = np.linspace(0,16,100)
y = x + np.random.randn(100)
plt.plot(x, y)

plt.title("Max temperatureF data")
plt.xlabel("Days")
plt.ylabel("TemperatureF")

plt.show()
```

__ Adding a legend __ 

```python
x = np.linspace(0,16,100)
y = x + np.random.randn(100)
plt.plot(x, y, label = "Temperature") # adding a label for the legend

plt.title("Max temperatureF data")
plt.xlabel("Days")
plt.ylabel("TemperatureF")

plt.legend(loc = "best")

plt.show()
```

__Change the x ticks rotation__

```python
plt.xticks(rotation = 90)
```

__Changing the x and y limits __

```python
plt.xlim(min(x), max(x))
```

   __2. Object-oriented way__

__Creating the plot__

```python
x = np.linspace(0,16,100)
y = x + np.random.randn(100)

fig = plt.figure(num = None, figsize = (8,4))  # num is to identify the figure, if None, then a new figure is created
ax = fig.add_subplot(111)   #row-column-number

ax.plot(x, y, linewidth = 3, label = "Temp")

plt.show()
```

__Labels, title, legend __

```python
x = np.linspace(0,16,100)
y = x + np.random.randn(100)
temp = pd.DataFrame({"Days": x, "Temps": y})

fig = plt.figure(num = None, figsize = (8,4))  # num is to identify the figure, if None, then a new figure is created
ax = fig.add_subplot(111)   #row-column-number

ax.plot(x, y, linewidth = 3, label = "Temp")

ax.set_xlabel("Days")
ax.set_ylabel("Temps")
ax.legend(loc = "best")
ax.set_title("Temps in April")
ax.annotate("My birthday", xy=(temp.iloc[50].Days,temp.iloc[50].Temps), xytext=(7, 13),
            arrowprops = dict(arrowstyle = '->', facecolor = 'black' , connectionstyle = 'arc3,rad=0'))
ax.text(11, 0, 'Some red box over here', style='italic',
        bbox={'facecolor':'red', 'alpha':0.5, 'pad':10})


fig.suptitle("Temperatures",  fontsize=14, fontweight='bold')
fig.subplots_adjust(top=0.85)  # add space between figure title and subplots


plt.show()
```
